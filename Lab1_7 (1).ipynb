2. Сбалансировано ли соотношение мужчин и женщин в наборе данных?

````python
men = df[df['Gender'] == 'Male']
women = df[df['Gender'] == 'Female']
print(len(men))
print(len(women))
print(len(men) / len(women) < 1.1 and len(men) / len(women) > 0.9)
#отношение близко к единице значит сбалансировано
````

3. Правда ли, что люди с высоким уровнем стресса имеют более высокое кровяное давление, чем те, кто не подвержен стрессу

````python
high_stress = df[df['Stress Levels'] > 8]
low_stress = df[df['Stress Levels'] < 2]
split_high = high_stress['Blood Pressure (s/d)'].str.split('/', expand=True)
split_low = low_stress['Blood Pressure (s/d)'].str.split('/', expand=True)
split_high[1] = pd.to_numeric(split_high[1])
split_high[0] = pd.to_numeric(split_high[0])
split_low[1] = pd.to_numeric(split_low[1])
split_low[0] = pd.to_numeric(split_low[0])
print((split_high[0].sum() + split_high[1].sum()) / len(split_high) - (split_low[0].sum() + split_low[1].sum()) / len(split_low))
````

Разница в давлении есть но незначительная

4. Правда ли, что уровень холестерина у мужчин и женщин отличается?

````python
men_cholesterol = df[df['Gender'] == 'Male']['Cholesterol Level (mg/dL)']
women_cholesterol = df[df['Gender'] == 'Female']['Cholesterol Level (mg/dL)']
print(men_cholesterol.mean())
print(women_cholesterol.mean())
print(men_cholesterol.mean() != women_cholesterol.mean())
````

Уровень холестерина почти одинаковый

5. Правда ли, что с возрастом женщины имеют свойство оценивать свои когнитивные способности хуже, чем в юности (порог юности - 30 лет, женщина в возрасте - 60 лет+)

````python
old_women_cognitive = df[(df['Age (years)'] >= 60) & (df['Gender'] == 'Female')]['Cognitive Function']
young_women_cognitive = df[(df['Age (years)'] < 30) & (df['Gender'] == 'Female')]['Cognitive Function']
print(old_women_cognitive.mean())
print(young_women_cognitive.mean())
print(old_women_cognitive.mean() < young_women_cognitive.mean())
````

6. Влияет ли физическая активность на вес в возрасте более 50 лет?

````python
old_people = df[df['Age (years)'] > 50]
high_level = old_people[old_people['Physical Activity Level'] == 'High']
moderate_level = old_people[old_people['Physical Activity Level'] == 'Moderate']
low_level = old_people[old_people['Physical Activity Level'] == 'Low']
print(high_level['Weight (kg)'].mean())
print(moderate_level['Weight (kg)'].mean())
print(low_level['Weight (kg)'].mean())
````

7. Сравните минимальный вес мужчин с высоким и низким заработком

````python
min_weight_rich = df[(df['Income Level'] == 'High') & (df['Gender'] == 'Male')]['Weight (kg)'].min()
min_weight_poor = df[(df['Income Level'] == 'Low') & (df['Gender'] == 'Male')]['Weight (kg)'].min()
print(min_weight_rich)
print(min_weight_poor)
````

8. Правда ли, что у людей с индексом массы тела больше 75го перцентиля встречаются хронические заболевания чаще, чем у людей с индесом массы тела в переделах от 30 до 60 перцентиля?

````python
percentile_75 = df['BMI'].quantile(0.75)
percentile_30 = df['BMI'].quantile(0.3)
percentile_60 = df['BMI'].quantile(0.6)
high_bmi_df = df[df['BMI'] > percentile_75]
medium_bmi_df = df[(df['BMI'] < percentile_60) & (df['BMI'] > percentile_30)]
high_bmi_count = pd.notna(high_bmi_df['Chronic Diseases']).sum() / len(high_bmi_df)
medium_bmi_count = pd.notna(medium_bmi_df['Chronic Diseases']).sum() / len(medium_bmi_df)
print(high_bmi_count)
print(medium_bmi_count)
print(high_bmi_count > medium_bmi_count)
````

9. Правда ли, что люди с высоким уровнем дохода имеют больший уровень физической активности, чем люди с низким доходом

````python
rich = df[df['Income Level'] == 'High']
poor = df[df['Income Level'] == 'Low']
rich_physical = rich['Physical Activity Level'].replace({'Low': 0, 'Moderate': 1, 'High': 2})
poor_physical = poor['Physical Activity Level'].replace({'Low': 0, 'Moderate': 1, 'High': 2})
print(rich_physical.mean())
print(poor_physical.mean())
print(rich_physical.mean() > poor_physical.mean())
````

10. Какая доля мужчин с высокой частотой пребывания на солнце имеет высокий заработок?

````python
sunny_men = df[(df['Gender'] == 'Male') & (df['Sun Exposure'] > 5)]
print(len(sunny_men[sunny_men['Income Level'] == 'High']) / len(sunny_men))
````
